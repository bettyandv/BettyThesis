```python
import pandas as pd
```

```python
data=pd.read_excel('betty_data/AF_AnaforeFeature_betty_processing.xls',sheet_name='Sheet2')
```

```python
data.columns=list(map(lambda x: str(x).split(' ')[0],data.columns))
```

```python
data=data[:-1]
```

```python
data.tail()
```

```python
result=pd.DataFrame()
for k,v in data.iterrows():
    stock_df=pd.DataFrame({'Stkcd':v['Stkcd'],'Date':v.index[1:],'Value':v[1:]})
    result=pd.concat((result,stock_df))
```

```python
result.to_excel('betty_data/processed_data.xls')
```
